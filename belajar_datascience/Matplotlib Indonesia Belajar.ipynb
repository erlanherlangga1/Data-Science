```python
# Multiple Sub Plot
x1 = np.linspace(0.0,5.0,100) # nilai min dan max bersifat inklusif, paramter ketiga menunjukkan jumlah nilai 
x2 = np.linspace(0.0,2.0,100)

y1 = np.cos(2 * np.pi * x1)
y2 = np.cos(2 * np.pi * x2)

x1
```

```python
import matplotlib
import matplotlib.pyplot as plt
from matplotlib import style
```

```python
def gambar_barplot():
    # Grouped bar plot

    data1 = [25,85,75,40,60]
    data2 = [40,35,20,55,20]
    kategori=['A','B','C','D','E']

    x=np.arange(len(kategori))
    x

    width = 0.3
    plt.bar(x-width/2,data1,width,label='Data 1') # parameter width merupakan lebar dari bar nya
    # nilai paramater pertama merupakan nilai index dari kategori
    plt.bar(x+width/2,data2,width,label='Data 2')

    plt.xticks(x,kategori) # xticks digunakan untuk melakukan pengaturan di point sumbu x dengan A, B, C
    # jika tidak dipakai maka akan muncul index 0,1,2,3,4,5

    plt.legend()


    plt.grid(linestyle='--',
            linewidth=1,
            axis='y', # grid nya berada di sumbu x
            alpha=0.75)

    plt.xlabel('kategori')
    plt.ylabel('jumlah')
    plt.title('Contoh Grouped Bar Plot')

    plt.show()
    
gambar_barplot()
```

# Box Plot

```python
np.random.seed(2)
data = np.random.normal(loc=100, scale=10,size=200)  # loc = nilai tengah, scale =mengatur sebaran data/ standar deviasi, size= mengatur jumlah datapoint yg dihasilkan
data
```

```python
plt.boxplot(data, labels=['Data'])

plt.title('Simple Box Plot')
plt.xlabel('sumbu x')
plt.ylabel('sumbu y')

plt.grid()
plt.show()

# garis tengah oren merupakan median
# garis kotak atas merupakan q3 dan bawah merupakan q1. q1 =median
# garis di atas merupakan nilai tertinggi dalam ambang batas 1.5 IQR, dan garis bawah merepresentasikan nilai terendah dari 1.5iqr ambang bawah
# marker bulat bulat merupakan data data outlier yang merupakan 1.5 dari nilai iqr
```

```python
plt.boxplot(data, labels=['Data'], notch=True)

plt.title('Notched Box Plot')
plt.xlabel('sumbu x')
plt.ylabel('sumbu y')

plt.grid()
plt.show()
```

```python
# pengaturan marker untuk outlier
plt.boxplot(data, labels=['Data'],
           showfliers=True,
           flierprops={'markerfacecolor':'r','marker':'d'})

plt.title('Simple Box Plot')
plt.xlabel('sumbu x')
plt.ylabel('sumbu y')

plt.grid()
plt.show()
```

```python
plt.boxplot(data, labels=['Data'], vert=False)

plt.title('Simple Box Plot')
plt.xlabel('sumbu x')
plt.ylabel('sumbu y')

plt.grid()
plt.show()
```

# Twin Axes

```python
x = np.linspace(1,10,25) # 25 merupakan jumlah data yang akan dihasilkan
x
```

```python
fig, ax1 = plt.subplots()

ax1.plot(x, np.exp(x),'bs-',label='exp')
ax1.set_xlabel('Sumbu X')
ax1.set_ylabel('exp')

ax2 = ax1.twinx()
ax2.plot(x,np.log(x),'ro-',label='log')
ax2.set_ylabel('log')

fig.suptitle('Contoh Twin Axes')
fig.legend(loc='upper left')

plt.show()
```

```python
# Kasus Twin Axes 2
x = np.random.randint(16,40, size=30)
x
```

```python
def C2F(celsius=0):
    return (celsius*1.8) + 32

def konversi_sumbu(ax1):
    y1, y2 = ax1.get_ylim()
    ax2.set_ylim(C2F(y1),C2F(y2))
    ax2.figure.canvas.draw()
    
fig, ax1 = plt.subplots()
ax2 = ax1.twinx()

ax1.callbacks.connect('ylim changed', konversi_sumbu)
ax1.plot(x)

ax1.set_xlabel('Hari')
ax1.set_ylabel('Celsius')
ax2.set_ylabel('Fahrenheit')

fig.suptitle('Temperature udara')
plt.show()
```

# Pengaturan Figure Layouts dengan GridSpec 

```python
import matplotlib.gridspec as gridspec
```

```python
# simple 2x2 layout dengan subplots
siswa = ['Cecep', 'Joki',' Wati','Ani']
uas = [68,83,96,72]

fig, axes = plt.subplots(ncols=2,nrows=2,constrained_layout=True)

axes[0][0].bar(siswa,uas)
axes[0][1].pie(uas,labels=siswa)
axes[1][0].pie(uas,labels=siswa,explode=[0,0,0,0.2])
axes[1][1].barh(siswa,uas)

plt.show()
```

```python
# simple 2x2 layout dengan GridSpec
siswa = ['Cecep', 'Joki',' Wati','Ani']
uas = [68,83,96,72]

fig = plt.figure(constrained_layout=True)
gs=gridspec.GridSpec(ncols=2,nrows=2,figure=fig)

ax1 = fig.add_subplot(gs[0,0])
ax2 = fig.add_subplot(gs[0,1])
ax3 = fig.add_subplot(gs[1,0])
ax4 = fig.add_subplot(gs[1,1])

ax1.bar(siswa,uas)
ax2.pie(uas,labels=siswa)
ax3.pie(uas,labels=siswa,explode=[0,0,0,0.2])
ax4.barh(siswa,uas)

plt.show()
```

```python
# Multiple Line Plot
ax = plt.gca() # akan mengembalikan objek akses (get current access)
df.plot(kind='line',x='nama',y='n_anak',color='green',ax=ax)
df.plot(kind='line',x='nama',y='n_pets',color='red',ax=ax)

plt.title('Line Plot')
plt.ylabel('Jumlah')

plt.show()
```
