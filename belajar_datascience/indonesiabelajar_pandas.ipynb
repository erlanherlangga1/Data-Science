# Cummulative Count pada Pandas Data Frame cumcount()

```python
df = pd.DataFrame({
    'penjual':['bejo','tejo','wati','bejo','cecep','tejo','wati','bejo'],
    'barang':['monitor','monitor','keyboard','mouse','keyboard','monitor','laptop','monitor']
})
df
```

```python
# mengenal kumulatif count dengan cumcount() (penjumlahan baris atau kemunculan nilai)
df['count tiap penjual'] = df.groupby('penjual').cumcount() + 1 
df
# + 1 digunakan unutk dimulai dari 1
```

```python
df['count tiap barang'] = df.groupby('barang').cumcount() + 1 
df
```

```python
df['count pasangan kolom'] = df.groupby(['barang','penjual']).cumcount() + 1 
df
```

# Membuat kolom baru dengan for loop dan f-string 

```python
df = pd.DataFrame({
    'penjual':['bejo','tejo','wati','bejo','cecep','tejo','wati','bejo'],
    'barang':['monitor','monitor','keyboard','mouse','keyboard','monitor','laptop','monitor']
})
df
```

```python
cols=['penjual','barang']

for col in cols:
    df[f'count_tiap_{col}'] = df.groupby(col).cumcount()+1

df
```

# Membuat kategori baru menurut threshold
