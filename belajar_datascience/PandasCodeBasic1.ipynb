# Handle Missing Data: fillna, dropna, interpolate

```python
df = pd.read_csv('data_csv/weather_data.csv',parse_dates=['day'])

type(df.day[0]) #cek jenis data
df.set_index('day',inplace=True)
df
```

```python
new_df = df.fillna(0)
new_df
```

```python
new_df = df.fillna({
    'temperature':0,
    'windspeed':0,
    'event':'no event'
})
new_df
```

```python
new_df = df.fillna(method='ffill') # mengisi nan value dengan value sebelumnya
new_df
```

```python
new_df = df.fillna(method='ffill', limit =1) # mengisi nan value dengan value sebelumnya hanya sekali
new_df
```

```python
new_df = df.fillna(method='bfill') # mengisi value dengan value sesudahnya
new_df
```

```python
new_df = df.fillna(method='bfill', axis='columns') # mengisi value dengan value kolom di samping kanannya atau secara horizontal
new_df
```

```python
new_df = df.interpolate()
new_df
# menggunakan nilai linier
```

```python
new_df = df.interpolate(method='time')
new_df
```

```python
new_df = df.dropna()
new_df
```

```python
new_df = df.dropna(how='all')
new_df
```

```python
new_df = df.dropna(thresh=1) # menghapus baris yang tidak memiliki value sama sekali, dan membiarkan baris yang memiliki minimal 1 value
new_df
```
