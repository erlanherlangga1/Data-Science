# Axes labels, Legend, Grid

```python
days=[1,2,3,4,5,6,7]
max_t=[32,51,52,48,47,49,46]
min_t=[43,42,40,44,33,35,37]
avg_t=[45,48,48,46,40,42,41]
```

```python
plt.xlabel('days')
plt.ylabel('Temperature')
plt.title('Weather')

# untuk memasukkan legend, kita perlu memasukkan argumen label ke dalam plt.plot

plt.plot(days,max_t,label='Max')
plt.plot(days,min_t,label='Min')
plt.plot(days,avg_t,label='Avg')

plt.legend(loc='best',shadow=True)
```

```python
plt.xlabel('days')
plt.ylabel('Temperature')
plt.title('Weather')

# untuk memasukkan legend, kita perlu memasukkan argumen label ke dalam plt.plot

plt.plot(days,max_t,label='Max')
plt.plot(days,min_t,label='Min')
plt.plot(days,avg_t,label='Avg')

plt.legend(loc='best',shadow=True)
plt.grid()
```

#  Bar Chart

```python
company=['GOOGL','AMZN','MSFT','FB']
revenue=[90,136,89,27]
profit=[40,2,34,12]
```

```python
ypos = np.arange(len(company))
ypos
```

```python
plt.xticks(ypos,company)
plt.ylabel('revenue{bln}')
plt.title('US Tech Stocks')
plt.bar(ypos,revenue,label='Revenue')
plt.legend()
```

```python
plt.xticks(ypos,company)
plt.ylabel('revenue{bln}')
plt.title('US Tech Stocks')
plt.bar(ypos-0.2,revenue,label='Revenue',width=0.4)
plt.bar(ypos+0.2,profit,label='Profit',width=0.4)
plt.legend()
```

```python
plt.xticks(ypos,company)
plt.title('US Tech Stocks')
plt.barh(ypos-0.2,revenue,label='Revenue')
plt.barh(ypos+0.2,profit,label='Profit')
plt.legend()
```

# Histogram

```python
blood_sugar = [113, 85, 90, 150, 149, 88, 93, 115, 135, 80, 77, 82, 129]
plt.hist(blood_sugar, rwidth=0.8) # by default number of bins is set to 10
# rwidth merupakan jarak antara bins 1 dengan lainnya
```

```python
# bins parameter
plt.hist(blood_sugar,rwidth=0.5,bins=4)
```

```python
plt.xlabel("Sugar Level")
plt.ylabel("Number Of Patients")
plt.title("Blood Sugar Chart")

plt.hist(blood_sugar, bins=[80,100,125,150], rwidth=0.9, color='g')
```

```python
blood_sugar_men = [113, 85, 90, 150, 149, 88, 93, 115, 135, 80, 77, 82, 129]
blood_sugar_women = [67, 98, 89, 120, 133, 150, 84, 69, 89, 79, 120, 112, 100]

plt.xlabel('sugar range')
plt.ylabel('freq')
plt.title('blood sugar analysis')
plt.hist([blood_sugar_men,blood_sugar_women], bins=[80,100,125,150], rwidth=0.95, color=['green','red'],label=['men','women'])
plt.legend()
```

```python
blood_sugar_men = [113, 85, 90, 150, 149, 88, 93, 115, 135, 80, 77, 82, 129]
blood_sugar_women = [67, 98, 89, 120, 133, 150, 84, 69, 89, 79, 120, 112, 100]

plt.xlabel('sugar range')
plt.ylabel('freq')
plt.title('blood sugar analysis')
plt.hist([blood_sugar_men,blood_sugar_women], bins=[80,100,125,150], rwidth=0.95, color=['green','red'],label=['men','women'],
        orientation='horizontal')
plt.legend()
```

# Pie Chart

```python
exp_vals = [1400,600,300,410,250]
exp_labels = ["Home Rent","Food","Phone/Internet Bill","Car ","Other Utilities"]
plt.pie(exp_vals,labels=exp_labels)
plt.show()
```

```python
plt.axis('equal')
```

```python
exp_vals = [1400,600,300,410,250]
exp_labels = ["Home Rent","Food","Phone/Internet Bill","Car ","Other Utilities"]
plt.pie(exp_vals,labels=exp_labels,autopct='%0.2f%%')
plt.show()
```

```python
exp_vals = [1400,600,300,410,250]
exp_labels = ["Home Rent","Food","Phone/Internet Bill","Car ","Other Utilities"]
plt.pie(exp_vals,labels=exp_labels,autopct='%0.0f%%',shadow=True,explode=[0.2,0,0,0,0])
# explode untuk memecah bagiannya
plt.show()
```

```python
exp_vals = [1400,600,300,410,250]
exp_labels = ["Home Rent","Food","Phone/Internet Bill","Car ","Other Utilities"]
plt.pie(exp_vals,labels=exp_labels,autopct='%0.0f%%',shadow=True,explode=[0.2,0,0,0,0],startangle=45)
# start angle untuk merotasi posisi dari pie chart yang dibuat
plt.show()
```
